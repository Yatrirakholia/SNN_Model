```python
from brian2 import *
```

```python
time_constant = 10*ms
threshold = -50*mV
reset = -65*mV # voltage
resting_potential = -65*mV
R_m = 100*Mohm
```

```python
eqs = '''
dv/dt = (resting_potential-v)/time_constant + (I/time_constant) :volt
I : amp #Input current
'''

eqs1 = '''
dv/dt = (resting_potential - v) / time_constant + (I * R_m) / time_constant : volt
I : amp  # Input current (in Amperes)
'''
```

```python
neuron = NeuronGroup(1, eqs1, threshold='v > threshold', reset='v = reset', method='exact')
# 1 is number of neuron
# eqs equation used
#
```

```python
# Set initial conditions
neuron.v = resting_potential
neuron.I = 1 * nA  # Constant input current
```

```python
# Monitor voltage and spikes
mon = StateMonitor(neuron, 'v', record=True)
spikemon = SpikeMonitor(neuron)
```

```python
# Run the simulation for 100ms
run(100 * ms)
```

```python
# Plot results
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))

# Plot membrane potential over time
plt.subplot(2,1,1)
plt.plot(mon.t / ms, mon.v[0] / mV)
plt.xlabel('Time (ms)')
plt.ylabel('Membrane potential (mV)')
plt.title('Neuron Voltage Over Time')

# Plot spike times
plt.subplot(2,1,2)
plt.plot(spikemon.t / ms, [1] * len(spikemon.t), 'ro')
plt.xlabel('Time (ms)')
plt.ylabel('Spikes')
plt.title('Neuron Spikes')
plt.ylim(0.5, 1.5)

plt.tight_layout()
plt.show()
```
